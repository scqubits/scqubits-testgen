```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import scqubits as scq
```

```python
N = 4    # arbitrary last index of the particular eigenstate to be used for comparison, i.e., use 0, 1, ..., N-1
M = 12   # number of evals to be compared
```

# Transmon qubit

```python
CPB = scq.Transmon(
    EJ=30.02,
    EC=1.2,
    ng=0.3,
    ncut=31
    )
```

```python
CPB.filewrite('./data/transmon-test.hdf5')
```

```python
TESTNAME = 'transmon_1.hdf5'

CPB.eigenvals(evals_count=M, filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'transmon_2.hdf5'

evals, evecs = CPB.eigensys(evals_count=N,  filename='./data/' + TESTNAME)
```

```python
# TESTNAME = 'transmon_3'

ng_list = np.linspace(-2, 2, 100)
CPB.plot_evals_vs_paramvals('ng', ng_list, evals_count=5, subtract_ground=False)
```

```python
TESTNAME = 'transmon_4.hdf5'

ng_list = np.linspace(-2, 2, 50)
tst = CPB.get_spectrum_vs_paramvals('ng', ng_list, evals_count=4, subtract_ground=False, get_eigenstates=True, filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'transmon_5.hdf5'

nmat = CPB.matrixelement_table('n_operator', evecs=None, evals_count=10, filename='./data/' + TESTNAME)
```

```python
# TESTNAME = 'transmon_6'

CPB.ng = 0.3
ej_vals = CPB.EJ * np.cos(np.linspace(-np.pi/2, np.pi/2, 40))
CPB.plot_evals_vs_paramvals('EJ', ej_vals, evals_count=4, subtract_ground=False)
```

```python
# TESTNAME = 'transmon_7'
    
CPB.plot_n_wavefunction(esys=None, which=1, mode='real')
```

```python
# TESTNAME = 'transmon_8'

CPB.plot_phi_wavefunction(esys=None, which=(0,2,6), mode='real');
CPB.plot_phi_wavefunction(esys=None, which=1, mode='abs_sqr');
```

```python
# TESTNAME = 'transmon_9'

CPB.plot_matrixelements('n_operator', evals_count=10);
```

```python
# TESTNAME = 'transmon_10'

mat_data = CPB.matrixelement_table('n_operator')
```

```python
# TESTNAME = 'transmon_11'

fig, ax = CPB.plot_matelem_vs_paramvals('n_operator', 'ng', ng_list, select_elems=4, filename='./data/test')
```
