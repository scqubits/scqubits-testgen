```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import scqubits as qubit
import scqubits.utils.plotting as plot

import numpy as np
```

```python
N = 4    # arbitrary last index of the particular eigenstate to be used for comparison, i.e., use 0, 1, ..., N-1
M = 12   # number of evals to be compared
```

# Flux qubit

```python
# parameters for the flux qubit
RATIO = 60.0
ALPHA = 0.8

flux_qubit = qubit.FluxQubit(
    EJ1 = 1.0, 
    EJ2 = 1.0, 
    EJ3 = ALPHA*1.0, 
    ECJ1 = 1.0/RATIO, 
    ECJ2 = 1.0/RATIO, 
    ECJ3 = 1.0/ALPHA/RATIO, 
    ECg1 = 50.0/RATIO, 
    ECg2 = 50.0/RATIO, 
    ng1 = 0.0, 
    ng2 = 0.0, 
    flux = 0.1, 
    ncut = 10,
)
```

```python
TESTNAME = 'fluxqubit_1.hdf5'

flux_qubit.eigenvals(evals_count=M, filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'fluxqubit_2.hdf5'

evals, evecs = flux_qubit.eigensys(evals_count=N,  filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'fluxqubit_3'

flux_list = np.linspace(0.45, 0.55, 50)
flux_qubit.plot_evals_vs_paramvals('flux', flux_list, evals_count=5, subtract_ground=False)
```

```python
TESTNAME = 'fluxqubit_4.hdf5'

flux_list = np.linspace(0.45, 0.55, 50)
tst = flux_qubit.get_spectrum_vs_paramvals('flux', flux_list, evals_count=4, subtract_ground=False, get_eigenstates=True, filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'fluxqubit_5.hdf5'

nmat = flux_qubit.matrixelement_table('n_1_operator', evecs=None, evals_count=10, filename='./data/' + TESTNAME)
plot.print_matrix(abs(nmat))
```

```python
TESTNAME = 'fluxqubit_6'

ej_vals = flux_qubit.EJ1 * np.cos(np.linspace(-np.pi/2, np.pi/2, 40))
flux_qubit.plot_evals_vs_paramvals('EJ1', ej_vals, evals_count=4, subtract_ground=False)
```
