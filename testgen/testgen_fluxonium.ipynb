# Initialization

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import scqubits as qubit
import scqubits.utils.plotting as plot

import numpy as np
```

```python
N = 4    # arbitrary last index of the particular eigenstate to be used for comparison, i.e., use 0, 1, ..., N-1
M = 12   # number of evals to be compared
```

# Fluxonium qubit

$H_\text{fl}=-4E_\text{C}\partial_\phi^2-E_\text{J}\cos(\phi-\varphi_\text{ext}) +\frac{1}{2}E_L\phi^2$

```python
fluxonium = qubit.Fluxonium(
    EJ = 8.9,
    EC = 2.5,
    EL = 0.5,
    flux = 0.33,
    cutoff = 110
)
```

```python
fluxonium
```

```python
TESTNAME = 'fluxonium_1.hdf5'

fluxonium.eigenvals(evals_count=M, filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'fluxonium_2.hdf5'

evals, evecs = fluxonium.eigensys(evals_count=N,  filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'fluxonium_3'

flux_list = np.linspace(-0.5, 0.5, 51)
fluxonium.plot_evals_vs_paramvals('flux', flux_list, evals_count=5, subtract_ground=False)
```

```python
TESTNAME = 'fluxonium_4.hdf5'

flux_list = np.linspace(-0.5, 0.5, 51)
tst = fluxonium.get_spectrum_vs_paramvals('flux', flux_list, evals_count=4, subtract_ground=False, get_eigenstates=True, filename='./data/' + TESTNAME)
```

```python
TESTNAME = 'fluxonium_5.hdf5'

nmat = fluxonium.matrixelement_table('n_operator', evecs=None, evals_count=10, filename='./data/' + TESTNAME)
plot.print_matrix(abs(nmat))
```

```python
TESTNAME = 'fluxonium_6'

ej_vals = fluxonium.EJ * np.cos(np.linspace(-np.pi/2, np.pi/2, 40))
fluxonium.plot_evals_vs_paramvals('EJ', ej_vals, evals_count=4, subtract_ground=False)
```
